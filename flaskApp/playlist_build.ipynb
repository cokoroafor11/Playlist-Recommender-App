```python
#Editorial
editorial_df_cleaned = editorial_df[['Playlist','Genres','Followers','Spotify Playlist URL']]
```

```python
editorial_df_cleaned.dtypes
```

```python
#Function to convert numbers: i.e. 10k to 10,000
#Shoutout github: https://stackoverflow.com/questions/39684548/convert-the-string-2-90k-to-2900-or-5-2m-to-5200000-in-pandas-dataframe
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x or 'k' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x or 'm' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x or 'b' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
editorial_df_cleaned['Followers'] = editorial_df_cleaned['Followers'].apply(value_to_float)
editorial_df_cleaned['Followers'] = editorial_df_cleaned['Followers'].astype('int')
```

```python
editorial_df_cleaned = editorial_df_cleaned[editorial_df_cleaned['Followers'] > 5000]
```

```python
editorial_df_cleaned
```

```python
independent_df_cleaned = independent_df[['Playlist','Genres','Followers','Spotify Playlist URL']]
```

```python
independent_df_cleaned.dtypes
```

```python
independent_df_cleaned['Followers'] = independent_df_cleaned['Followers'].apply(value_to_float)
independent_df_cleaned['Followers'] = independent_df_cleaned['Followers'].astype('int')
```

```python
independent_df_cleaned = independent_df_cleaned[independent_df_cleaned['Followers'] > 5000]
```

```python
independent_df_cleaned
```

```python
#Merge dfs together
all_playlist_df = pd.concat([editorial_df_cleaned,independent_df_cleaned])
```

```python
all_playlist_df.dropna(subset = ['Followers'],inplace = True)
all_playlist_df
```

```python
#Download csv of all playlists
all_playlist_df.to_excel('all_spotify_playlists.xlsx',index = False)
files.download('all_spotify_playlists.xlsx')
```
